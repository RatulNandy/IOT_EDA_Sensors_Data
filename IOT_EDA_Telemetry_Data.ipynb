```python
# Importing necessary libraries

import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Customizing figure features

plt.style.use('seaborn-paper')
sns.set(rc={"figure.dpi":500, 'savefig.dpi':500})
plt.rcParams.update(plt.rcParamsDefault) 
```

```python
#Setting the filepath and importing the dataset

os.chdir("/Users/ratulnandy/Documents/MSAAI/IoT&DataAnalytics/Module4")
df = pd.read_csv('iot_telemetry_data.csv')
```

#### Inference -  It can be seen that the dataset contains 405184 records with 9 columns. The columns are time stamp, device name, carbon monooxide %, humidity %, light detected, liquid petroleum %, motion detected, smoke % and temperature  

```python
# Checking the number of unique devices 

df['device'].unique()
```

#### Inference - It is observed that there are three unique devices

```python
# Changing the device name to meaningful names (as device number). Going forward we'll use the column 
# devicenumber to track the device

df['devicenumber'] = df['device'] 
df.devicenumber.replace(['00:0f:00:70:91:0a', '1c:bf:ce:15:ec:4d', 'b8:27:eb:bf:9d:51'], [1, 2, 3], inplace=True)
```

```python
df.head(5)
```

#### The column 'ts' is a continuous representation of time. Changing the field's datetime stamp

```python
# The column 'time_stamp' will be used to track the time

df['time_stamp'] = pd.to_datetime(df['ts'],unit='s')
```

```python
df.head(5)
```

```python
# Heatmap to visualize missing values in the dataset

fig, ax = plt.subplots(figsize=(50,20)) 
sns.heatmap(df.isnull())
plt.show()
```

```python
df. isnull(). sum()
```

#### Inference - It is observed that there are no null values in the dataset

## Checking Correlation between the data ( Heat Map )

```python
# Checking the correlation between carbon monooxide, humidity, lpg, smoke and temperature


df_conditions = df[['co','humidity','lpg','smoke','temp']].copy()
df_conditions.head(5)
```

```python
df_cor = df[['co','humidity','lpg','smoke','temp','devicenumber']].copy()
```

### Plotting the correlated data

### CO Vs LPG

```python
plt.figure(figsize=(8,6))
dev = ['Device 1', 'Device 2', 'Device 3']
scatter = plt.scatter(df_cor.co, 
            df_cor.lpg,
            s=150,
            c=df_cor.devicenumber.astype('category').cat.codes)
plt.xlabel("Carbon Monoxide")
plt.ylabel("LPG")
plt.legend(handles=scatter.legend_elements()[0], 
           labels=dev,
           title="Devices")
plt.show()
```

### CO Vs Smoke

```python
plt.figure(figsize=(8,6))
dev = ['Device 1', 'Device 2', 'Device 3']
scatter = plt.scatter(df_cor.co, 
            df_cor.smoke,
            s=150,
            c=df_cor.devicenumber.astype('category').cat.codes)
plt.xlabel("Carbon Monoxide")
plt.ylabel("Smoke")
plt.legend(handles=scatter.legend_elements()[0], 
           labels=dev,
           title="Devices")
plt.show()
```

### LPG Vs Smoke


```python
plt.figure(figsize=(8,6))
dev = ['Device 1', 'Device 2', 'Device 3']
scatter = plt.scatter(df_cor.lpg, 
            df_cor.smoke,
            s=150,
            c=df_cor.devicenumber.astype('category').cat.codes)
plt.xlabel("LPG")
plt.ylabel("Smoke")
plt.legend(handles=scatter.legend_elements()[0], 
           labels=dev,
           title="Devices")
plt.show()
```

### Checking correlation between 'Temperature' and 'Humidity'

```python
# Checking Temperature vs Humidy

plt.figure(figsize=(8,6))
dev = ['Device 1', 'Device 2', 'Device 3']
scatter = plt.scatter(df_cor.temp, 
            df_cor.humidity,
            s=150,
            c=df_cor.devicenumber.astype('category').cat.codes)
plt.xlabel("Temperature F")
plt.ylabel("Humidity %")
plt.legend(handles=scatter.legend_elements()[0], 
           labels=dev,
           title="Devices")
plt.show()
```

#### Inference - From the plot, we can observe that for Device 3, the temperature and humidy are quite stabe in the temperature range (20F-25F) and humidy(40%-60%). For Device 1, there is a lot of fluctuation in the humidity at around 20F. For Device 2, there is no specific pattern

### Analysis of Light & Motion Detection 

```python
# Creating a separate dataframe

df_1_cat=df_1[['light','motion','time_stamp']].copy()
df_2_cat=df_2[['light','motion','time_stamp']].copy()
df_3_cat=df_3[['light','motion','time_stamp']].copy()
```

```python
df_1_cat["light"] = df_1_cat["light"].astype(int)
df_1_cat["motion"] = df_1_cat["motion"].astype(int)

df_2_cat["light"] = df_2_cat["light"].astype(int)
df_2_cat["motion"] = df_2_cat["motion"].astype(int)

df_3_cat["light"] = df_3_cat["light"].astype(int)
df_3_cat["motion"] = df_3_cat["motion"].astype(int)
```

### Plotting Light Detection

```python
plt.plot(df_1_cat.time_stamp, df_1_cat.light)
plt.plot(df_2_cat.time_stamp, df_2_cat.light)
plt.plot(df_3_cat.time_stamp, df_3_cat.light)
plt.legend(['Device 1', 'Device 2', 'Device 3'])
plt.xlabel('Time')
plt.ylabel('Light Detection')
plt.title('Light Detection Captured by the device')
plt.xticks(rotation=90)
plt.show()
```

#### Inference - It can be observed that Device 2 is placed in a location where there is a constant presence of light. Device 3 is placed in a location which had few occurrence of light detection towards the earlier timespan. Later on there was no light detected. Device 1 is situated in a location which gets light at regular intervals

### Plotting Motion Detection

```python
plt.plot(df_1_cat.time_stamp, df_1_cat.motion)
plt.legend(['Device 1'])
plt.xlabel('Time')
plt.ylabel('Motion Detection in Device 1')
plt.title('Motion Detection Captured by Device 1')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(df_2_cat.time_stamp, df_2_cat.motion)
plt.legend(['Device 2'])
plt.xlabel('Time')
plt.ylabel('Motion Detection in Device 2')
plt.title('Motion Detection Captured by Device 2')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(df_3_cat.time_stamp, df_3_cat.motion)
plt.legend(['Device 3'])
plt.xlabel('Time')
plt.ylabel('Motion Detection in Device 3')
plt.title('Motion Detection Captured by Device 3')
plt.xticks(rotation=90)
plt.show()
```

#### Inference - It can be seen that Device 1 is placed in a remote area with minimum motion detection. Device 2 is placed in a heavily trafficed area. Device 3 is placed in a moderately trafficed area
